```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("ggplot")

%matplotlib inline
```

### 1. Linear Function
Create a sample of points with some noise that follow the equation: $$Y = AX + B$$ where A = 2.5 and B = 20.

Now, plot it as either a cloud of points or a line.

```python
A = 2.5
B = 20
samples = 50 #number of samples

#equation: Y = AX + B + noise
X = np.linspace(-10,10,samples) #X goes from -10 to +10
Y = A * X + B + np.random.normal(0,5,samples)

#Scatter Plot
plt.scatter(X,Y)
plt.show()
```

### 2. Logistic Function
Represent the logistic, or sigmoid, function between -20 and 20. Remember its formula is:

$$\displaystyle S(x)={\frac {1}{1+e^{-x}}}={\frac {e^{x}}{e^{x}+1}}$$

_Hint_: you will need an X and a Y to plot against it. 

_Hint_: check out the function np.exp

```python
#Resolution
samples = 50

#equation: 1/(1+e**(-x))
X = np.linspace(-5,5,samples)
Y = 1/(1+np.exp(-X))

#Plot
plt.plot(X,Y)
plt.show()
```
