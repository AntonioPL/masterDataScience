Loading CSV

```python
airbus_records = pd.read_csv("./data/airbus_records.csv", delimiter = "|")
optd_aircraft = pd.read_csv("./data/optd_aircraft.csv", delimiter = "^")
optd_airlines = pd.read_csv("./data/optd_airlines.csv", delimiter = "^")
optd_por_public = pd.read_csv("./data/optd_por_public.csv", delimiter = "^")
```

Importing tables in database _optd_

```python
conn = sqlite3.connect('./data/bbdd/optd.db')
```

```python
airbus_records.to_sql("airbus_records", conn, if_exists='replace')
optd_aircraft.to_sql("optd_aircraft", conn, if_exists='replace')
optd_airlines.to_sql("optd_airlines", conn, if_exists='replace')
optd_por_public.to_sql("optd_por_public", conn, if_exists='replace')
```

__Querying__:

```python
query = "SELECT model,nb_engines FROM optd_aircraft WHERE nb_engines >= 4"
```

```python
pd.read_sql_query(query, conn)
```
